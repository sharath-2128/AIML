```
#ANSWER 10 CASE 1 
def  rev ( n ): 
    rev = 0 
    while  n > 0 : 
        rem = n % 10 
        rev = rev * 10 + rem 
        n = n // 10 
    return  rev
        
n = int ( input ( "enter a number:" )) 
s1 = n * n 
rev1 = rev ( n ) 
s2 = rev1 * rev1 
if  s1 == rev ( s2 ): 
    print ( "Adam Number" ) 
else : 
    print ( "Not a Adam Number" )
```
